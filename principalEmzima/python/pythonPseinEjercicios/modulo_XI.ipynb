Actividad 2
Desarrolla un programa que genere una matriz de tamaño 5x15. Tu tarea 
consiste en llenar la matriz con unos y ceros, donde los 1 ocupan el 
borde externo de la matriz y los 0 llenarán el área interior.

Por ejemplo, el aspecto final de tu matriz deberá ser el siguiente:
```python
	111111111111111
	100000000000001
	100000000000001
	100000000000001
	111111111111111
```

````python
import numpy as np

matriz_np = np.zeros((5,15),dtype=int) # con este metodo se rellena una matriz de 5x15 con 0
matriz_np [0,:] = 1    # la primera fila todas las columnas
matriz_np [:,0] = 1    # la primera columna todas las filas
matriz_np [:,-1] = 1   # la ultima columna todas las filas
matriz_np [-1,:] = 1   # la ultima fila todas las columnas
df = pd.DataFrame(matriz_np)  
display(df)
````

````python
# otra forma

matriz_np = np.ones((5,15),dtype=int) # con este metodo se rellena una matriz de 5x15 con 1
matriz_np [1:-1,1:-1] = 0    # se excluyen la primera y la ultima filas, la primera y la ultima columnas
df = pd.DataFrame(matriz_np)  
display(df)
````

````python
# con list comprehensive
matriz= [[1 for _ in range(15)]for _ in range(5) ]   # crear matriz con list comprehensive
matriz_np = np.array(matriz)   # pero para la etapa siguiente convertimos a objeto np
matriz_np[1:-1,1:-1] = 0
matriz_np
df = pd.DataFrame(matriz_np)  
display(df)
````

````python
# con una funcion como en pseint pero no vale la pena
def crear_matriz(numRow,numCol,matriz):
    for i in range(numRow):
        for j in range(numCol):
            if (i==0 or (j==0 or j==numCol-1)) or (i==numRow-1 or (j==0 or j==numCol-1)):
                matriz[i][j] = 1
            else:
                matriz[i][j] = 0 

matriz= [[2 for _ in range(15)]for _ in range(5) ] # pero es importante esta parte para que la funcion pueda operar sobre posiciones que ya existen
crear_matriz(5,15,matriz)
matriz
df = pd.DataFrame(matriz)  
display(df)
````

Una distribuidora de Nescafé cuenta con 4 representantes que recorren toda 
Argentina para ofrecer sus productos. Para la gestión administrativa, el país 
está dividido en cinco zonas: Norte, Sur, Este, Oeste y Centro. Mensualmente, 
la empresa registra los datos de ventas de los representantes en cada 
zona y recopila diversas estadísticas sobre su desempeño.

Se requiere un programa que lea el monto de las ventas de los 
representantes en cada zona y realice los siguientes cálculos:
* Total de ventas de una zona especificada por el usuario.
* Total de ventas de un representante seleccionado por el usuario en cada una de las zonas.
* Total de ventas de todos los representantes.

````python
from IPython.display import clear_output
def segmento(text,arreglo):
    choice = ''
    while choice not in ['S','s']:
        print('*'*(len(text)+11))
        print(f'MODULO POR {text}')
        print('*'*(len(text)+11))
        print('Opciones')
        for i,name in enumerate(arreglo):
            print(f'{i}- {name}')
        print('S- Salir')
        choice = input('Selecciona alguna de las opciones anteriores')
        if choice in [str(number) for number in range(len(arreglo))]:
            # en esta parte seleccionar si va a imprimir las columnas (zonas) y por eso df[nombre de la columna]
            # los metodos to_frame() para convertir una serie (tipo dato de pandas) y .T transponer como explicamos antes
            # en el caso de una fila se debe usar df.loc[nombre de la fila] por eso depende de si es zona o representantes
            display(df[arreglo[int(choice)]].to_frame().T) if text == 'ZONAS' else display(df.loc[arreglo[int(choice)]].to_frame().T)
            if input('Desea salir S/N') in ['S','s']:
                clear_output(wait=True)
                break
            else:
                clear_output(wait=True)

repres_ventas = ['Carlos','Juan','Ana','Laura']
zonas = ['Norte','Sur','Este','Oeste','Centro']
filas, columnas = 4,5
matriz= [[rn.randint(10,100) for _ in range(columnas)]for _ in range(filas) ]
df = pd.DataFrame(matriz, index=repres_ventas, columns=zonas)
df.loc['Total Zona'] = df.sum(axis=0)    # suma de las columnas de cada fila y añadir fila al final 
df['Total Vendedor'] = df.sum(axis=1)    # suma de las filas de cada columna y añadir columna al final 
is_salir = False
while not is_salir:
    print('Opciones sobre el Total de ventas de ')
    print('1- Una zona especifica con cada representante y el total ')
    print('2- Un representante en cada zona y el total')
    print('3- Todos los representantes')
    print('4- Salir')
    choice = input('Selecciona alguna de las opciones anteriores ')
    match choice:
        case '1':
            clear_output(wait=True)
            segmento('ZONAS',zonas)
        case '2':
            clear_output(wait=True)
            segmento('REPRESENTANTES DE VENTAS',repres_ventas)
        case '3':
            clear_output(wait=True)
            display(df)
        case '4':
            is_salir = True
            print('Hasta la vista baby')
        case _:
            print('Che bolodo entre lo que esta escrito')
````

Complementario 3
Una empresa de venta de productos por correo desea realizar una 
estadística de las ventas de cada uno de sus productos a lo largo de una 
semana, distribuyendo los 5 productos en los 5 días hábiles. Se desea conocer:
* El total de ventas por cada día de la semana.
* El total de ventas de cada producto a lo largo de la semana.
* El producto más vendido en cada semana.
* El nombre, el día de la semana y la cantidad del producto más vendido.
El informe final tendrá el siguiente formato:
```python
				Lunes  Martes  Miércoles  Jueves  Viernes  Total producto
	Producto 1						
	Producto 2						
	Producto 3						
	Producto 4						
	Producto 5						
	Total semana						
	Producto más vendido
```	

````python
# Hecho con list comprehension todos los datos menos el index del mas vendido
productos = ['Cafe','Te','Azucar','Yerba','Agua','Total']
dias_semana = ['Lunes','Martes','Miercoles','Jueves','Viernes','Total Producto']
filas, columnas = 5,5
matriz= [[rn.randint(10,100) for _ in range(columnas)]for _ in range(filas) ] # matriz de datos aleatorios
matriz = [fila +[sum(fila)] for fila in matriz] # suma de las filas de cada columna y añadido al final como columna total producto
fil_col_sum = [sum(fila[i] for fila in matriz) for i in range(columnas+1)] # suma de las columnas de cada fila
matriz.append(fil_col_sum) # y añadido al final como fila Total
df = pd.DataFrame(matriz, columns=dias_semana, index=productos) # index define los valores de la columna de referencia a la izquierda
df.loc['Mas vendido'] = df.iloc[:-1].idxmax() # el metodo idxmax() devuelve el index del maximo valor de cada columna
df
````
