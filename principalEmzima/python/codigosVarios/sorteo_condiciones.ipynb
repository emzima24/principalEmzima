```python
import random

# Lista de números
cantidad_us = 22
numeros = list(range(1, cantidad_us+1))

# Número de personas
lista_personas = ['Fran','Milton','Roger','Ana','David', 'Biss','Mar','Stef','Cin','Angi','Edw','Emi']
lista_personas.remove('Emi')
# Barajar los números para asignación aleatoria
random.shuffle(numeros)

# Asignar los números a las personas
asignacion = {persona: [] for persona in lista_personas}

# Calcular la cantidad base de números que recibe cada persona
cantidad_base = len(numeros) // len(lista_personas)
sobrantes = len(numeros) % len(lista_personas)

# Asignar los números de manera equitativa
indice = 0
for persona in lista_personas:
    for _ in range(cantidad_base):
        asignacion[persona].append(numeros[indice])
        indice += 1

lista_personas_sobran = lista_personas.copy()
#lista_personas_sobran.remove('Emi')
# Asignar los números sobrantes a algunas personas
if sobrantes > 0:
    for i in range(sobrantes):
        asignacion[lista_personas_sobran[i]].append(numeros[indice])
        indice += 1

# Imprimir la asignación
for persona, nums in asignacion.items():
    print(f'{persona}: {nums}')
```

```python
import random
import pandas as pd

sorteo = True

# Lista de números
restrictions = {
    'Angelica': [40, 41, 42],
    'Bisset': [24, 27, 53, 54, 55],
    'Cintia': [8, 9],
    'David': [5, 6, 22, 29],
    'Francisco': [3, 13, 14, 20, 21],
    'Milton': [11, 12, 25, 38, 39],
    'Stefi': [15, 16, 17, 18, 19],
    'Ana': [46, 47, 48, 49, 56],
    'Edwin': [7, 51, 52, 57, 58],
    'Martin': [28, 35, 10],
    'Florencia': [30, 31, 32],
    'Fabian': [4, 26, 43],
    'Silvia': [23],
    'Juan': [50],
    'Andres': [5, 6, 22, 29],
    'Sofia': [15, 16, 17, 18, 19],
    'Micaela': [28, 35, 10],
    'Roxana': [],
    'Emiliano': [],
}
all_numbers = list(range(3, 59))

def assign_numbers(restrictions, all_numbers):
    # Inicializar el diccionario de asignaciones
    assignments = {el: [] for el in restrictions.keys()}

    # Lista de números disponibles para cada elemento
    available_numbers = {el: [num for num in all_numbers if num not in restrictions[el]] for el in restrictions.keys()}

    # Asignar 1 número a
    for el in ['Angelica', 'Cintia','Stefi','Sofia','Micaela']:
        if available_numbers[el]:
            num = random.choice(available_numbers[el])
            assignments[el].append(num)
            all_numbers.remove(num)
            for k in available_numbers:
                if num in available_numbers[k]:
                    available_numbers[k].remove(num)

    # Asignar 3 números a los demás, excepto Emiliano
    for _ in range(3):
        for el in restrictions.keys():
            if el not in ['Angelica', 'Cintia','Stefi','Sofia','Micaela','Emiliano']:
                if available_numbers[el]:
                    num = random.choice(available_numbers[el])
                    assignments[el].append(num)
                    all_numbers.remove(num)
                    for k in available_numbers:
                        if num in available_numbers[k]:
                            available_numbers[k].remove(num)

    # Asignar todos los números restantes a Emiliano
    assignments['Emiliano'] = all_numbers

    return assignments

assignments = assign_numbers(restrictions, all_numbers.copy())

if sorteo:
    df = pd.DataFrame(assignments.values(), index=assignments.keys())
else:
    df = pd.DataFrame(columns=range(0,12), index=restrictions.keys())
df = df.fillna(0).astype(int)
df.replace(0,'-',inplace=True)
display(df)

count_distinct = (df != '-').sum().sum()
print(f'La cantidad de numeros asignados es {count_distinct}')
```

```python
import random
import pandas as pd

sorteo = True
reasingacion =  {
    'Cintia': [15],
    'Milton': [18],
    'Martin': [57],
    'Florencia': [58, 16, 55],
    'Fabian': [27],
    'Silvia': [21,17]    
}

change = [5,23,25,29,28,36,38,42,46,40]
len(change)

# Lista para almacenar el resultado final de las reasignaciones
final_reasingacion = {}

# Crear una copia de la lista de números para evitar modificar la original
remaining_change = change.copy()

# Barajar la lista de números a cambiar para aleatoriedad
random.shuffle(remaining_change)

print(remaining_change)

# Asignar números al azar de la lista 'change' a cada clave en 'reasingacion'
for key in reasingacion.keys():
    num_to_assign = len(reasingacion[key])  # número de elementos a asignar
    assigned_numbers = [remaining_change.pop() for _ in range(num_to_assign)]
    final_reasingacion[key] = assigned_numbers

if sorteo:
    df = pd.DataFrame(final_reasingacion.values(), index=final_reasingacion.keys())
else:
    df = pd.DataFrame(columns=range(0,3), index=final_reasingacion.keys())
df = df.fillna(0).astype(int)
df.replace(0,'-',inplace=True)
display(df)
```
