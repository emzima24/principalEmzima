```python
import random
import pandas as pd

# Lista de números
restrictions = {
    'Monica': {36,37,38,39,48,58,86,87,91,93,94,95,96,97,98,99,100,101},  
    'Adriana': {65,66,67,68,74,75,76,78,79,80,81,82,83,84,85},   
    'Juan': {50,51,60,61,70,112},  
    'Carina': {41,52,53,55,56,57,62,64,88,92,104},  
    'Silvia': {69,71,72,73,77,117},
    'Raúl': {105,106,107,108,109,110,111,115,116},
    'Darimar': {118,119,120,121,122,123,124,125,126,127,128,129,130},
    'Emiliano': {131,132,133,134,135,136,137,138,139,140},  
}
all_numbers = [value for sublist in restrictions.values() for value in sublist]
all_numbers.sort()

def assign_numbers(restrictions,all_numbers):
    # Inicializar el diccionario de asignaciones
    assignments = {el:[] for el in restrictions.keys()}
   
    # Lista de números disponibles para cada elemento
    available_numbers = {el: [num for num in all_numbers if num not in restrictions[el]] for el in restrictions.keys()}

    for el in restrictions.keys():
        if len(available_numbers[el]) < 11:
            raise ValueError(f"No hay suficientes números disponibles para el elemento {el} con las restricciones dadas.")
    
    for _ in range(11):
        for el in restrictions.keys():
            # Elegir un número aleatorio de los disponibles para el elemento
            num = random.choice(available_numbers[el])
            assignments[el].append(num)
            # Eliminar el número elegido de todos los conjuntos disponibles
            all_numbers.remove(num)
            for k in available_numbers:
                if num in available_numbers[k]:
                    available_numbers[k].remove(num)
    
    return assignments

assignments = assign_numbers(restrictions,all_numbers.copy())
#print(assignments)

df = pd.DataFrame(assignments.values(),index=assignments.keys())
df
```
